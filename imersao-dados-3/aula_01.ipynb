### Desafio 05: Procure na documentação o método 'query' do pandas

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html

```python
df_g0_greater_than_0_v2 = df.query('`g-0` > 0')

df_g0_greater_than_0_v2.head()
```

Nota: O acento crase para a coluna só foi necessário devido ao hífen. O método query não permite alguns caracteres na string, gerando erro. O hífen é uma delas. Esse problema foi resolvido usando o acento crase.
Para queries nas colunas como `id`, `tratamento`, `tempo`, `dose` ou `droga`, o uso da crase não seria necessário, pois não há caracteres não aceitos.


Agora vamos verificar se o resultado da busca com o uso da máscara - True ou False - foi o mesmo resultado com o uso do método `query`.

```python
# (df_g0_greater_than_0_v2 == df_g0_greater_than_0).any(1)
# pd.concat([df_g0_greater_than_0_v2, df_g0_greater_than_0]).drop_duplicates(keep=False)

df_g0_greater_than_0_v2.equals(df_g0_greater_than_0)
```

### Desafio 06: Deixar os gráficos com título, label - usar matplotlib.pyplot

### Desafio 07: Renomear as colunas tirando o hífen

```python
df.columns
```

```python
new_column_names = []

for column in df.columns:
    new_name = column.replace('-', '')
    new_column_names.append(new_name)
    
new_column_names[:10]
```

- Outra maneira: Usando list comprehension:

```python
new_column_names = [column.replace('-', '') for column in df.columns]

new_column_names[:10]
```

- Mudando o nome das colunas

```python
df.columns = new_column_names
```
