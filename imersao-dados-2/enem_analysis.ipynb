### Desafio 1:

Plotar as médias, medianas e moda nas notas de LC e MT. (botar a linha vertical)

```python
def box_hist_plot_com_media_mediana_moda_provas (prova):
    
    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, 
                                        gridspec_kw= {"height_ratios": (0.2, 1)}, 
                                        figsize=(12,7))

    media_nota = inscritos_nota_total_maiorq_zero[prova].mean()
    mediana_nota = inscritos_nota_total_maiorq_zero[prova].median()
    moda_nota = inscritos_nota_total_maiorq_zero[prova].mode()[0]

    sns.boxplot(x=inscritos_nota_total_maiorq_zero[prova],
                ax=ax_box, boxprops=dict(alpha=0.6))
    ax_box.axvline(media_nota, color='r', linestyle='--', lw=2)
    ax_box.axvline(mediana_nota, color='g', linestyle='-', lw=2)
    ax_box.axvline(moda_nota, color='y', linestyle='-', lw=2)

    sns.histplot(data=inscritos_nota_total_maiorq_zero, 
                 x=prova, ax=ax_hist, alpha=0.6)
    ax_hist.axvline(media_nota, color='r', linestyle='--', lw=2)
    ax_hist.axvline(mediana_nota, color='g', linestyle='-', lw=2)
    ax_hist.axvline(moda_nota, color='y', linestyle='-', lw=2)

    plt.legend({'Media':media_nota,'Mediana':mediana_nota,'Moda':moda_nota})

    ax_box.set(xlabel='')

    plt.show()
```

```python
box_hist_plot_com_media_mediana_moda_provas('NU_NOTA_LC')
```

```python
box_hist_plot_com_media_mediana_moda_provas('NU_NOTA_MT')
```

Estudar intervalo de confiança.

```python
def calcula_mse(resultado):

    media_resultado_score = (resultado['test_score']*-1).mean()
    desvio_padrao_resultado_score = (resultado['test_score']*-1).std()
    limite_inferior = media_resultado_score - (2*desvio_padrao_resultado_score)
    limite_superior = media_resultado_score + (2*desvio_padrao_resultado_score)


    print('Média: '+str(media_resultado_score))
    
    print('Desvio Padrão: '+str(desvio_padrao_resultado_score))
    
    print('Intervalo de confiança: \nde '+str(limite_inferior)+' até '+str(limite_superior))
```

```python
calcula_mse(resultado_10)
```

O cross validation não tem aleatoriedade, ou seja, ele vai pegar, agrupar pelos x primeiros, os x próximos e assim por diante.

Se o nosso dataset está ordenado, isso não é legal, por exemplo, ordenado por estado, um estado pode ficar de fora do treino.

Então, antes de aplicar cross validation, podemos embaralhar os dados.

```python
from sklearn.model_selection import KFold
```

```python
# shuffle=True vai embaralhar, então sem ele o resultado será o mesmo
partes = KFold(n_splits=10, shuffle=True)

modelo_arvore = DecisionTreeRegressor(max_depth=3)

cross_validate(modelo_arvore, notas_entrada, notas_saida, 
               scoring='neg_mean_squared_error')

resultado_kfold = cross_validate(modelo_arvore, notas_entrada, notas_saida, 
                                 scoring='neg_mean_squared_error', cv=partes)

calcula_mse(resultado_kfold)
```

mas se quisermos sempre reproduzir um mesmo resultado para fazer algum tipo de avaliação, então devemos setar algum random_state

```python
import numpy as np
```

```python
seed = 1232
np.random.seed(seed)


partes = KFold(n_splits=10, shuffle=True)

modelo_arvore = DecisionTreeRegressor(max_depth=3)

cross_validate(modelo_arvore, notas_entrada, notas_saida, 
               scoring='neg_mean_squared_error')

resultado_kfold = cross_validate(modelo_arvore, notas_entrada, notas_saida, 
                                 scoring='neg_mean_squared_error', cv=partes)

calcula_mse(resultado_kfold)
```

Analisar a mudança do max_depth

```python
def regressor_arvore(nivel):
    
    seed = 1232
    np.random.seed(seed)
    
    partes = KFold(n_splits=10, shuffle=True)

    modelo_arvore = DecisionTreeRegressor(max_depth=nivel)

    cross_validate(modelo_arvore, notas_entrada, notas_saida, 
                   scoring='neg_mean_squared_error')

    
    # return_train_score=True retorna não só os dados de teste como os de treino também
    resultado_kfold = cross_validate(modelo_arvore, notas_entrada, notas_saida, 
                                     scoring='neg_mean_squared_error', cv=partes,
                                     return_train_score=True)
    
    print(f'Treino: {(resultado_kfold["train_score"]*-1).mean()} | Teste: {(resultado_kfold["test_score"]*-1).mean()}')
```

```python
regressor_arvore(4)
```

```python
for i in range(1,20):
    regressor_arvore(i)
```

Analisando somente os dados de teste: Quanto mais níveis, profundidade melhor? Não, a partir do 8 começo ter um aumento do mse, o que não é bom.

Enquanto que para os dados de treino melhora o mse quando aumenta a profundidade.

O que isso significa? O nosso modelo está ficando muito bom para prever os nossos dados de treino, ou seja, eles estão ficando acostumados com esses dados, decorando eles. E na hora de generalizar para os dados de teste, ele não consegue generalizar de maneira adequada. Ou seja, decorou os dados de treino e não generalizou o modelo para os dados de teste.
